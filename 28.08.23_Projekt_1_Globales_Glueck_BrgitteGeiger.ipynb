```python


# Daten aus der CSV-Datei laden
lucky_country2023 = pd.read_csv("WHR2023.csv")

# LÃ¤nderranking erstellen
ranking = lucky_country2023.sort_values(by='Ladder score', ascending=False)

ranking = ranking.head(10)

# Donut-Plot erstellen
plt.pie(ranking["Ladder score"], labels=ranking["Country name"], autopct="%1.1f%%", startangle=90)
plt.axis("equal") # KreisfÃ¶rmige Darstellung sicherstellen
plt.title("Die Top 10 der glÃ¼cklichsten LÃ¤nder 2023")
# plt.legend()

# Diagramm anzeigen
plt.show()
```

```python
# Erstellen von Top-10- und Bottom-10-Datenrahmen und VerknÃ¼pfen dieser Datenrahmen
top10 = lucky_country2023.set_index('Country name')['Ladder score'].nlargest(10).to_frame()
bottom10 = lucky_country2023.set_index('Country name')['Ladder score'].nsmallest(10).to_frame()
dfconcat = pd.concat([top10, bottom10], axis=0)
dfconcat = dfconcat.sort_values(by="Ladder score")

fig = plt.figure(figsize=(10,8))
cmap = plt.cm.RdYlGn
norm = plt.Normalize(vmin=dfconcat['Ladder score'].min(), vmax=dfconcat['Ladder score'].max())
colors = [cmap(norm(value)) for value in dfconcat['Ladder score']]

g = plt.barh(y=dfconcat.index, width=dfconcat["Ladder score"], color=colors)
plt.ylabel("LÃ¤nder")
plt.xlabel("GlÃ¼ckpunktzahl")
plt.title("Die 10 glÃ¼cklichsten und am wenigsten glÃ¼cklichsten LÃ¤nder 2023", fontsize = 20)

plt.show()
```

```python
# Sehen Sie sich die Verteilung der GlÃ¼ckswerte an, indem Sie Mittelwerte (rot) und Medianwerte (schwarz) einzeichnen

plt.figure(figsize=(10,8))
plt.title("Verteilung der GlÃ¼ckpunktzahl(Ranking) der Top 10 LÃ¤nder 2023", fontsize = 20)
plt.xlabel('GlÃ¼ckspunktzahl(Ranking)')
plt.ylabel('Anzahl der LÃ¤nder')
plt.legend('Top 10 LÃ¤nder')
plt.hist(top10, bins = np.arange(0.2,10, 0.5), edgecolor = 'white')

plt.axvline(x = top10['Ladder score'].mean(), color = 'red')
plt.axvline(x = top10['Ladder score'].median(), color = 'black')
```

```python
import pandas as pd
import numpy as np
import seaborn as sns

# Lassen Sie uns SchÃ¤tzungen der Kerndichte fÃ¼r die 6 Faktoren erstellen

fig = plt. figure()
fig, axes = plt.subplots(2, 3, figsize= (10, 6))
plt.tight_layout(pad= 3)

plta = sns.kdeplot( data=lucky_country2023, x='Logged GDP per capita',color='#37F090',fill=True, ax=axes[0,0])
pltb = sns.kdeplot( data=lucky_country2023, x='Healthy life expectancy',color='#00D0A3',fill=True, ax=axes[0,1])
pltc = sns.kdeplot( data=lucky_country2023, x='Social support',color='#00ADA7',fill=True, ax=axes[0,2])
pltd = sns.kdeplot( data=lucky_country2023, x='Freedom to make life choices',color='#008A9A',fill=True, ax=axes[1,0])
plte = sns.kdeplot( data=lucky_country2023, x='Generosity',color='#08687E',fill=True, ax=axes[1,1])
pltf = sns.kdeplot( data=lucky_country2023, x='Perceptions of corruption',color='#2F4858',fill=True, ax=axes[1,2])

plta.axvline(x = lucky_country2023['Logged GDP per capita'].mean(), color = 'red')
plta.axvline(x = lucky_country2023['Logged GDP per capita'].median(), color = 'black')
pltb.axvline(x = lucky_country2023['Healthy life expectancy'].mean(), color = 'red')
pltb.axvline(x = lucky_country2023['Healthy life expectancy'].median(), color = 'black')
pltc.axvline(x = lucky_country2023['Social support'].mean(), color = 'red')
pltc.axvline(x = lucky_country2023['Social support'].median(), color = 'black')
pltd.axvline(x = lucky_country2023['Freedom to make life choices'].mean(), color = 'red')
pltd.axvline(x = lucky_country2023['Freedom to make life choices'].median(), color = 'black')
plte.axvline(x = lucky_country2023['Generosity'].mean(), color = 'red')
plte.axvline(x = lucky_country2023['Generosity'].median(), color = 'black')
pltf.axvline(x = lucky_country2023['Perceptions of corruption'].mean(), color = 'red')
pltf.axvline(x = lucky_country2023['Perceptions of corruption'].median(), color = 'black')
pltf.axvline(x = lucky_country2023['Perceptions of corruption'].median(), color = 'black')

plt.show()
```
